### Figure 1

```python
import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig1.npy' )
plt.figure()
plt.plot(a,color='black',marker='o')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

### Figure 2

```python
b=np.load('data-fig2.npy')
x,y,z=list(b[:,0]),list(b[:,1]),list(b[:,2])
plt.plot(x,color='red',label='Red')
plt.plot(y,color='green',label='Green')
plt.plot(z,color='blue',label='Blue')
plt.xlabel('X value')
plt.ylabel('Y value')
plt.legend()
plt.show
```

### Figure 3

```python
c=np.load('data-fig3.npy')
plt.scatter(c[:,0],c[:,1],color='b',alpha=0.3,s=200)
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show
```

### Figure 4

```python
d=np.load('data-fig4.npy')
D=['A','B','C','D','E']
plt.bar(D,d[:],color='silver')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

### Figure 5

```python
e=np.load('data-fig5.npy')
plt.pcolormesh(e,cmap='RdBu',vmax=1.0,vmin=-1.0)
plt.colorbar()
plt.show()
```
