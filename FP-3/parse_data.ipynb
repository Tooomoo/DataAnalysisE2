# parse_data.ipynb

This notebook parses the data files used for the FP-2 assignment. 

<br>
<br>

First let's read the attached data file:

```python
import pandas as pd

df = pd.read_csv('BeijingPM20130101_20151231.csv')

df.describe()
```

<br>
<br>

The dependent and independent variables variables (DVs and IVs) that we are interested in are:

**DVs**:
- PM2.5 concentration (μg/m)

**IVs**:
- Dew_Point_Temperature　("DEWP" column in the CSV file)
- Humidity ("HUMI" column in the CSV file)
- Pressure  ("PRES" column in the CSV file)
- Temperature("TEMP" column in the CSV file)
- Precipitation("precipitation" column in the CSV file)
- Hour("hour" column in the CSV file)

<br>
<br>

Let's extract the relevant columns:

```python

df = df[['hour', 'PM_US Post', 'DEWP', 'HUMI', 'PRES', 'TEMP','precipitation']]

df.describe()
```

<br>
<br>

Next let's use the `rename` function to give the columns simpler variable names:

```python
df = df.rename( columns={'PM_US Post':'PM_US_Post', 'hour':'Hour', 'DEWP':'Dew_Point_Temperature', 'HUMI':'Humidity', 'PRES':'Pressure', 'TEMP':'Temperature', 'precipitation':'Precipitation'} )

df.describe()
```
