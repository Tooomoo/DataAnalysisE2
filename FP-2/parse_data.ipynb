The dependent and independent variables variables (DVs and IVs) that we are interested in are:

**DVs**:
- PM2.5 concentration (g/m)

**IVs**:
- Dew point temperature("DEWP" column in the CSV file)
- Humidity ("HUMI" column in the CSV file)
- Pressure  ("PRES" column in the CSV file)
- Temperature("TEMP" column in the CSV file)
- Precipitation("precipitation" column in the CSV file)


<br>
<br>

Let's extract the relevant columns:

```python

df = df0[  ['DEWP', 'HUMI', 'PRES', 'TEMP','precipitation']  ]

df.describe()
```

<br>
<br>

Next let's use the `rename` function to give the columns simpler variable names:

```python
df = df.rename( columns={'DEWP':'dew point temperature', 'HUMI':'humidity', 'PRES':'pressure', 'TEMP':'temperature'} )

df.describe()
```
